## a. Why is multiplicative seasonality necessary for this series?

As we can see below, when we subtract trailing twelve months mean from each period, residuals are increse through the years. But when we divide each period by TTM mean, residuals stay constant through the years.

```python
twelve_month_mean = sales.rolling(3).mean()
```

```python
(sales - twelve_month_mean).plot(figsize=(15,7))
```

```python
(sales / twelve_month_mean).plot(figsize=(15,7))
```

## b. Apply Holt-Wintersâ€™ multiplicative method to the data. Experiment with making the trend damped.

```python
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
```

```python
train_length = int(len(sales)*2/3)
test_length = len(sales) - train_length
train_sales = sales[0:train_length]
test_sales = sales[train_length:]
```

```python
fit1 = Holt(train_sales).fit(smoothing_level=0.8, smoothing_slope=0.2, optimized=False)
fcast1 = fit1.forecast(test_length).rename("Holt's linear trend")

fit2 = Holt(train_sales,exponential=True).fit(smoothing_level=0.8, smoothing_slope=0.2, optimized=False)
fcast2 = fit2.forecast(test_length).rename("Exponential trend")

fit4 = Holt(train_sales,exponential=True).fit(smoothing_level=0.7, smoothing_slope=0.2, optimized=False)
fcast4 = fit4.forecast(test_length).rename("Exponential trend")

fit3 = Holt(train_sales, damped=True).fit(smoothing_level=0.8, smoothing_slope=0.2)
fcast3 = fit3.forecast(test_length).rename("Additive damped trend")

ax = sales.plot(color="gray", figsize=(16,8))
#fit1.fittedvalues.plot(ax=ax, color='blue')
fcast1.plot(ax=ax, color='blue',  legend=True)
#fit2.fittedvalues.plot(ax=ax, color='red')
fcast2.plot(ax=ax, color='red',  legend=True)
fcast4.plot(ax=ax, color='navy',  legend=True)
#fit3.fittedvalues.plot(ax=ax, color='green')
fcast3.plot(ax=ax, color='yellow', legend=True)

plt.show()
```

```python
fit1 = ExponentialSmoothing(train_sales, seasonal_periods=12, trend='add', seasonal='add').fit(use_boxcox=True)
fit2 = ExponentialSmoothing(train_sales, seasonal_periods=12, trend='add', seasonal='mul').fit(use_boxcox=True)
fit3 = ExponentialSmoothing(train_sales, seasonal_periods=12, trend='add', seasonal='add', damped=True).fit(use_boxcox=True)
fit4 = ExponentialSmoothing(train_sales, seasonal_periods=12, trend='add', seasonal='mul', damped=True).fit(use_boxcox=True)
results=pd.DataFrame(index=[r"$\alpha$",r"$\beta$",r"$\phi$",r"$\gamma$",r"$l_0$","$b_0$","SSE"])
params = ['smoothing_level', 'smoothing_slope', 'damping_slope', 'smoothing_seasonal', 'initial_level', 'initial_slope']
results["Additive"]       = [fit1.params[p] for p in params] + [fit1.sse]
results["Multiplicative"] = [fit2.params[p] for p in params] + [fit2.sse]
results["Additive Dam"]   = [fit3.params[p] for p in params] + [fit3.sse]
results["Multiplica Dam"] = [fit4.params[p] for p in params] + [fit4.sse]

ax = sales.plot(figsize=(10,6), color='gray', title="Forecasts from Holt-Winters' multiplicative method" )
ax.set_ylabel("Sales")
ax.set_xlabel("Period")
#fit1.fittedvalues.plot(ax=ax, style='--', color='red')
#fit2.fittedvalues.plot(ax=ax, style='--', color='green')

#fit1.forecast(test_length).rename('Holt-Winters (add-add-seasonal)').plot(ax=ax,  color='red', legend=True)
fit4.forecast(test_length).rename('Holt-Winters (add-mul-seasonal)').plot(ax=ax, color='green', legend=True)

plt.show()

results.round(3)
```
