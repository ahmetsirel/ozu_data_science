```python
from IPython.display import HTML, display

import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.sandbox.regression.predstd import wls_prediction_std
from sklearn.model_selection import train_test_split
from sklearn import metrics
from scipy import stats
import seaborn as sns

sns.set_style("darkgrid")
import pandas as pd
import numpy as np
import ast
import random
import math 
import time
import sys 
from sklearn.svm import LinearSVR
from sklearn.svm import SVR

np.random.seed(123)

df = pd.read_csv('WHOdata.csv', encoding='latin-1')

df = df[df["SMALL"]==0]

df= df.drop('CNAME', axis = 1)
df= df.drop('LCOMP', axis = 1)

df['COUNTRY'] = df.COUNTRY.astype('int')

df = pd.get_dummies(df)

print("The data size:", df.shape)

labels = np.array(df['COMP'])
df= df.drop('COMP', axis = 1)

train_x, test_x, train_y, test_y = train_test_split(df, labels, test_size = 0.25, random_state = 142)


train_x = sm.add_constant(train_x) 
test_x = sm.add_constant(test_x) 


reg1 = sm.OLS(train_y, train_x)
type(reg1)

results = reg1.fit()
type(results)

print(results.summary())
y_pred = results.predict(test_x)

print("train_x:", train_x.shape)
print("train_y:", train_y.shape)
print("test_x:", test_x.shape)
print("test_y:", test_y.shape)

# print (metrics.mean_squared_error(test_y, y_pred))
# print(np.sqrt(metrics.mean_squared_error(test_y, y_pred)))

# print(np.mean(np.abs((test_y - y_pred) / test_y)) * 100)
# print(np.max(np.abs((test_y - y_pred) / test_y)) * 100)
# print(np.mean(np.abs((test_y - y_pred))))
print(np.max(np.abs((test_y - y_pred))))
```
