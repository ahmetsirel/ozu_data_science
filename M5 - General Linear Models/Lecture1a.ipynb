```python
from statsmodels.formula.api import ols
from statsmodels.sandbox.regression.predstd import wls_prediction_std

from scipy import stats

import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
import ast
import random
import math 
import time
import sys 



df = pd.read_csv('cow_birth_weights.csv', encoding='latin-1')

Weight = df['Wt 0']

print("Sample Size:", len(Weight))
print("Mean:",Weight.mean())
print("Median:",Weight.median())
print("Std Dev:",Weight.std())
print("Std Err Mean:",stats.sem(Weight))
```

```python
Weight.hist(bins=58, range=(85,115),figsize=(15,15))

print(stats.skew(Weight))
```

```python
z_critical = stats.norm.ppf(q = 0.95)  # Get the z-critical value*

print("z-critical value:")              # Check the z-critical value
print(z_critical)                        

#margin_of_error = z_critical * stats.sem(Weight)
margin_of_error = z_critical * (Weight.std()/math.sqrt(len(Weight)))

confidence_interval = (Weight.mean() - margin_of_error,
                       Weight.mean() + margin_of_error)  

print("Confidence interval:")
print(confidence_interval)
```
