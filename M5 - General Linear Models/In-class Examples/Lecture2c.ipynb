```python
from IPython.display import HTML, display

import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.sandbox.regression.predstd import wls_prediction_std

from scipy import stats

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")

import pandas as pd

df = pd.read_csv('Advertising.csv', encoding='latin-1')

print("Sample Size:", len(df))
print("Sales Mean:", df["sales"].mean())
print("Sales Median:",df["sales"].median())
print("Sales Std Dev:",df["sales"].std())

print("Newspaper Mean:", df["newspaper"].mean())
print("Newspaper Median:",df["newspaper"].median())
print("Newspaper Std Dev:",df["newspaper"].std())

ax1 = df.plot.scatter(x='newspaper', y='sales', c='DarkBlue')
```

```python
df = sm.add_constant(df) 

reg1 = sm.OLS(endog=df['sales'], exog=df[['const','newspaper']], missing='drop')
type(reg1)

results = reg1.fit()
type(results)

print(results.summary())
```

```python
df_plot = df.dropna(subset=['sales', 'newspaper'])

# Plot predicted values

plt.scatter(df_plot['newspaper'], results.predict(), alpha=0.5, label='predicted')

# Plot observed values

plt.scatter(df_plot['newspaper'], df_plot['sales'], alpha=0.5, label='observed')

plt.legend()
plt.title('OLS predicted values')
plt.xlabel('newspaper')
plt.ylabel('sales')
plt.show()
```
