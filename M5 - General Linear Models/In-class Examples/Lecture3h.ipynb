```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn import metrics
import statsmodels.api as sm
from sklearn.model_selection import train_test_split

np.random.seed(12)

def main():
      
    #data = pd.io.stata.read_stata('keane.dta')
    #data.to_csv('keane.csv', index=False)
    data = pd.read_csv('keane.csv')

    data.dropna(inplace= True)
  
    data = data[data["year"]==87]
    
    #print(data.head)
    print ("Number of observations :: ", len(data.index))
    print ("Number of columns :: ", len(data.columns))
   # print ("Headers :: ", data.columns.values)
    
    
    data= data.drop('id', axis = 1)
    data= data.drop('numyrs', axis = 1)
    data= data.drop('year', axis = 1)
    data= data.drop('choice', axis = 1)
    data= data.drop('expwc', axis = 1)
    data= data.drop('expbc', axis = 1)
    data= data.drop('manuf', axis = 1)
    data= data.drop('enroll', axis = 1)
    data= data.drop('employ', axis = 1)
    data= data.drop('attrit', axis = 1)
    data= data.drop('expser', axis = 1)
    
    #data= data.drop('exper', axis = 1)
    #data= data.drop('expersq', axis = 1)
    
    data['const'] = 1
    print ("Headers: ", data.columns.values)
    
    labels = np.array(data['status'])
    features= data.drop('status', axis = 1)
    features = np.array(features)
    
    mlogit = sm.MNLogit(labels, features)
    fmlogit = mlogit.fit()
    fmlogit_margeff = fmlogit.get_margeff()
    print(fmlogit.params)
    print(fmlogit_margeff.summary())
    #print(fmlogit.summary())

if __name__ == "__main__":
    main()
```
